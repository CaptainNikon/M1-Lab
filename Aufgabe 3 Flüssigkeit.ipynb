```python
def gaus(x,a,x0,sigma):
        return a*np.exp(-(x-x0)**2/(2*sigma**2))

def calcthickness(mean,sigma,mean2,sigma2,array1,d1,d2,d3,d4,k,name,a=1):
    x1a = np.asarray(array1[d1:d2,0])
    x1e = np.asarray(array1[d3:d4,0])

    y1a = np.asarray(array1[d1:d2,1])
    y1e = np.asarray(array1[d3:d4,1])
    
    #fit gaussian
    popt1,pcov1 = curve_fit(gaus,x1a,y1a,p0=[a,mean,sigma])
    popt3,pcov3 = curve_fit(gaus,x1e,y1e,p0=[a,mean2,sigma2])

    plt.plot(x1a, gaus(x1a,*popt1), label='Gauß-Fit-Anfang 1')
    plt.plot(x1e, gaus(x1e,*popt3), label='Gauß-Fit-Ende 1')
    plt.scatter(array1[:,0],array1[:,1],c="lightskyblue",s=2,label="Daten")
    plt.legend(loc=2)
    plt.xlabel(r"Wellenvektor k $[\mathrm{cm^{-1}}]$")
    plt.ylabel("relative Intensität")
    plt.xlim(array1[d1,0]-30,array1[d4,0]+30)
    plt.ylim(array1[d1,1]-10,array1[d4,1]+10)
    plt.savefig("schichtdicke/del"+name+".pdf")

    n = 1.37927
    dk = popt3[1] - popt1[1]
    d = k/(2*n*dk)
    
    udk = np.sqrt((popt3[2])**2+(popt1[2])**2)
    ud = np.sqrt((udk*(-(k)/(2*n*dk**2)))**2)
    
    print("Schichtdicke= "+str(np.round(d*10000,2)),"; ","Messfehler= "+str(np.round(ud*10000,2)))
    print("Abstand der Maxima= "+str(np.round(dk,2)))
```

```python
def cos(x,a,b,c,d):
    return a*np.cos(b*x+c)+d

def calcthickness_cosine(pop1,pop2,array1,mean,sigma,mean1,sigma1,k,name,a=1):
    x = array1[:,0]
    y = array1[:,1]
    
    left_index = np.abs(x - (mean-sigma)).argmin()
    right_index = np.abs(x - (mean+sigma)).argmin()
    
    x1a = np.asarray(x[left_index:right_index])
    y1a = np.asarray(y[left_index:right_index])
    
    left_index1 = np.abs(x - (mean1-sigma1)).argmin()
    right_index1 = np.abs(x - (mean1+sigma1)).argmin()
    
    x1e = np.asarray(x[left_index1:right_index1])
    y1e = np.asarray(y[left_index1:right_index1])
    
    #fit gaussian
    popt1,pcov1 = curve_fit(cos,x1a,y1a,p0=[pop1])
    popt3,pcov3 = curve_fit(cos,x1e,y1e,p0=[pop2])
    
    leftpeakx = np.max(cos(x1a,*popt1))
    rightpeakx = np.max(cos(x1e,*popt3))
    
    leftpeak = np.interp(leftpeakx, y1a,x1a)
    rightpeak = np.interp(rightpeakx, y1e,x1e)
    
    plt.plot(x1a, cos(x1a,*popt1), label='Cos-Fit-Anfang 1')
    plt.plot(x1e, cos(x1e,*popt3), label='Cos-Fit-Ende 1')
    plt.scatter(array1[:,0],array1[:,1],c="lightskyblue",s=2,label="Daten")
    plt.legend(loc=2)
    plt.xlabel(r"Wellenvektor k $[\mathrm{cm^{-1}}]$")
    plt.ylabel("relative Intensität")
    plt.xlim(array1[left_index,0]-30,array1[right_index1,0]+30)
    plt.ylim(array1[left_index,1]-10,array1[right_index1,1]+10)
    #plt.savefig("schichtdicke/del"+name+".pdf")

    
    dk = np.abs(leftpeak-rightpeak)
    d = k/(2*n*dk)
    
    print(leftpeak,rightpeak)
    print("Schichtdicke= "+str(np.round(d*10000,3)))
    print("Abstand der Maxima= "+str(np.round(dk,2)))
```

```python
def gaus(x,a,x0,sigma):
        return a*np.exp(-(x-x0)**2/(2*sigma**2))

def peakstuff(mean,sigma,x,y,a=1):
    extinction1,ext1error = [],[]
    extinction2,ext2error= [],[]
    
    left_index = np.abs(x - (mean-sigma)).argmin()
    right_index = np.abs(x - (mean+sigma)).argmin()
    
    x1a = np.asarray(x[left_index:right_index])
    y1a = np.asarray(y[left_index:right_index])

    
    #fit gaussian
    popt1,pcov1 = curve_fit(gaus,x1a,y1a,p0=[a,mean,sigma])

    a_height,c_width = popt1[1], np.abs(popt1[2])

    peakindex1 = np.argmin(np.abs(_5iso_std[:,0]-a_height))

    p1a = np.argmin(np.abs(_5iso_std[:,0]-a_height+c_width/2))
    p1e = np.argmin(np.abs(_5iso_std[:,0]-a_height-c_width/2))

    for i in range(3,len(newd)):
        extinction1.append(source[i][peakindex1])
        ext1error.append(np.max([source[i][p1a:p1e]])-np.min([source[i][p1a:p1e]]))
        
    plt.plot(x1a, gaus(x1a,*popt1), label='Gauß-Fit-Anfang 1')
    plt.scatter(x,y,c="lightblue",s=2)
    plt.xlabel(r"Wellenvektor k $[\mathrm{cm^{-1}}]$",fontsize=12)
    plt.ylabel("Extinktion",fontsize=12)
    plt.xlim(x[left_index]-5,x[right_index]+5)
    plt.ylim(y[left_index]-10,y[right_index]+10)
    plt.show()
    
    plt.plot(newd[3:len(newd)], extinction1,label="Daten")
    plt.errorbar(newd[3:len(newd)], extinction1,fmt='o', xerr=newderr[3:len(newderr)],yerr= ext1error,markersize=2,capsize =2,label="Fehler")
    plt.xlabel(r"Schichtdicke $[\mathrm{\mu m}]$", fontsize=12)
    plt.ylabel(r"Extinktion [%]",fontsize=12)
    plt.title(r"Extinktion bei k $\approx$"+str(np.round(popt1[1],2))+ "$\mathrm{cm^{-1}}$")
    plt.legend(loc=0)
    plt.savefig("newplots/extinction1.pdf")
    plt.show()
```
